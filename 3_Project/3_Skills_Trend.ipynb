# In-Demand Skills Trending for Data Analysts

Methodology

2. Re-analyse based on total jobs percentage
3. Plot monthly skill demand

## Import Libraries & Data

```python
# import libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas() 
    
# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index = 'job_posted_month_no', columns = 'job_skills', aggfunc = 'size', fill_value = 0)
```

```python
df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()
```

```python
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending = False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

```python
# iloc[:, :5] returns all the rows and only the top 5 skills
df_DA_US_pivot.iloc[:, :5].plot()

plt.xlabel('Months')
plt.ylabel('Skill Count')
plt.title('Trending Skills for Data Analysts')
plt.show()
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_perc = df_DA_US_pivot.div(DA_totals / 100, axis = 0)

df_DA_US_perc
```

```python
df_DA_US_perc = df_DA_US_perc.reset_index()
df_DA_US_perc['job_posted_month'] = df_DA_US_perc['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format = '%m').strftime('%b'))
df_DA_US_perc = df_DA_US_perc.set_index('job_posted_month')
df_DA_US_perc = df_DA_US_perc.drop(columns = 'job_posted_month_no')

df_DA_US_perc
```

```python
df_plot = df_DA_US_perc.iloc[:, :5]

sns.lineplot(df_plot, dashes = False, markers = True, palette = 'tab10')
sns.set_theme(style = 'ticks')
sns.despine()

# label lines at the last data point
for col in df_plot.columns:
    y = df_plot[col].iloc[-1]
    x = df_plot.index[-1]
    plt.text(
        x, y,
        col,
        fontsize = 13,
        va = "center",
        ha = "left")

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals = 0))

plt.xlabel('2023')
plt.ylabel('Likelihood in Posting')
plt.title('Trending Top Skills for Data Analysts')
plt.legend().remove()

plt.show()
```

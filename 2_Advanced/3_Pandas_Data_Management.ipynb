```python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# everytime you run code, random records are fetched.
# random_state ensures you see the same records.
# use any number you want for random_state
df.sample(10, random_state = 5)
```

```python
# fetching original dataset with duplicates & all
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
# original is assigned to altered
df_altered = df_original
```

```python
df_altered.loc[:5, 'salary_year_avg']
```

```python
median_salary = df_altered['salary_year_avg'].median()

median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

df_altered.loc[:5, 'salary_year_avg']
```

```python
# df_original is supposed to have NaN values but it doesn't
# by assigning df_original to df_altered, whatever happens to
# dfa reflects on dfo
df_original.loc[:5, 'salary_year_avg']
```

```python
# this is why dataframes are exactly the same
print(id(df_original))
print(id(df_altered))
```

```python
# loading original dataset
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
# using .copy() copies data from df_original to df_altered
# whatever happens in dfa won't reflect in dfo
df_altered = df_original.copy()
```

```python
print('ID of df_original:             ', id(df_original))
print('ID of df_altered:              ', id(df_altered))
print('Are these dataframes the same? ', id(df_altered) == id(df_original))
```

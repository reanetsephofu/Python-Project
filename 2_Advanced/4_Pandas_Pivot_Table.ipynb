# Pivot Table

```python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### Counts of Job Titles

```python
# index is the column to group by
# values is columns you want to aggregate
# columns is similar to partition by in SQL
# aggfunc is the aggregation you want to perform
df.pivot_table(values = 'salary_year_avg', columns = 'job_country', index = 'job_title_short', aggfunc = 'median')
```

```python
# similar to code above. except index & columns values were swapped so
# table can be longer, and not wider. more readable
df.pivot_table(values = 'salary_year_avg', columns = 'job_title_short', index = 'job_country', aggfunc = 'median')
```

```python
# selects top 6 highest paying countries
top_countries = df['job_country'].value_counts().head(6).index

# new dataframe is created using above pivot table
df_job_country_salary = df.pivot_table(
    values = 'salary_year_avg',
    index = 'job_country',
    columns = 'job_title_short',
    aggfunc = 'median'
)

# .loc[top_countries] will filter pivot table to only show aforementioned top 6 countries
df_job_country_salary = df_job_country_salary.loc[top_countries]

# specific job titles you want to analyse
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

# this will filter to only show the aforementioned job titles
df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind = 'bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('Countries')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

## Exporting Data

```python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
    
# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# lowercase %b will return abbreviation of month name
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
# variable to store list of months
months = df['job_posted_month'].unique()

# dictionary to store jobs posted every month
dict_months = {month : df[df['job_posted_month'] == month] for month in months}

# extracts jobs posted on particular month
dict_months['Dec']
```

```python
# combines records from first quarter
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index = True)

df_q1
```

# Data Cleaning

```python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
# to fill out NaN values of salary_year_avg
median_salary_year_avg = df['salary_year_avg'].median() 
```

```python
# to fill out NaN values of salary_hour_avg
median_salary_hour_avg = df['salary_hour_avg'].median()
```

```python
# df_filled is duplicate of original df
# in case of mistake, original dataset isn't affected
df_filled = df

# replace NaN values in salary_year_avg & salary_hour_avg
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year_avg)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour_avg)
```

```python
df_filled[['salary_year_avg', 'salary_hour_avg']]
```

```python
# dataframe for duplicate values derives values from df_filled dataframe
df_unique = df_filled

# drop duplicates
df_unique = df_unique.drop_duplicates()

# compares original df with dropped-duplicate df
print('Length of original df:        ', len(df_filled))
print('Length of drop duplicates df: ', len(df_unique))
print('Rows dropped:                 ', len(df_filled) - len(df_unique))
```

```python
# drops duplicate job postings from same company
df_unique = df_unique.drop_duplicates(subset = ['job_title', 'company_name'])

print('Length of original df:        ', len(df_filled))
print('Length of drop duplicates df: ', len(df_unique))
print('Rows dropped:                 ', len(df_filled) - len(df_unique))
```

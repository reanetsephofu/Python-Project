```python
# import libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas() 
    
# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

#### Pie Chart

```python
# startangle determines where you want chart to start plotting
# autopct shows associated percentages
# % is compulsory
# .1 is amount of decimal places you want
# f means floating number
# %% shows actual percentage symbol
df['job_work_from_home'].value_counts().plot(kind = 'pie', startangle = 90, autopct = '%1.1f%%')

plt.title('Work From Home Status')
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind = 'pie', startangle = 180, autopct = '%1.1f%%')

plt.title('Job Titles')
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle = 180, autopct = '%1.1f%%', labels = ['False', 'True'])
    ax[i].set_title(title)

fig.tight_layout()

plt.show()
```

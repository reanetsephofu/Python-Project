```python
count = 1

while count < 5:
  print(count) # will print out all numbers less than 5. 5 not printed bc 5 = 5
  count += 1 # will keep adding 1 until value directly before 5 (4) is reached.
```

```python
years_experience = {'Reanetse': 3,
                    'Anele': 6,
                    'Tracey': 4,
                    'Martha': 3,
                    'Nalane': 7}

years = 0
index = 0

years_list = list(years_experience.items())

while years < 5: # will exclude values < 5, assess those >= 5
  key, years = years_list[index]
  if years >= 5: # assesses values >= 5 after those < 5 were removed
    print(key, "has", years, "years of experience")
  index += 1 # returns the first value that exceeds 5
```

## Exercise: Skill Investigation

Filter what jobs to apply for based on a list of skills

```python
my_skills = ['Python', 'SQL', 'Excel']

job_roles = [
    {'role': 'Data Analyst', 'skills': ['Python', 'SQL', 'Excel']},
    {'role': 'Data Scientist', 'skills': ['Python', 'R', 'Machine Learning', 'Deep Learning']},
    {'role': 'Machine Learning Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Scikit-Learn']},
    {'role': 'Data Engineer', 'skills': ['Python', 'Apache Spark', 'Hadoop', 'SQL']},
    {'role': 'Business Intelligence Analyst', 'skills': ['Python', 'SQL', 'Tableau', 'Power BI', 'Excel']},
    {'role': 'Quantitative Analyst', 'skills': ['R', 'Python', 'MATLAB', 'Statistics']},
    {'role': 'Operations Analyst', 'skills': ['Python', 'SQL', 'Data Visualization', 'Process Improvement']},
    {'role': 'Database Administrator', 'skills': ['SQL', 'Oracle', 'MySQL', 'Database Management']},
    {'role': 'AI Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Computer Vision']},
    {'role': 'Statistician', 'skills': ['R', 'SAS', 'Python', 'Statistical Modeling']}
]
```

Coding

* loop through the job_roles
* verify if all my_skills are inside each job_roles[skills]
* if true (or qualified), add to qualified list of jobs

```python
qualified_roles = []

for x in job_roles: # x stores the roles
  qualified = True

  for c in my_skills: # loops through skills i possess
    if c not in x['skills']: # checks if your skill exists in job posting
      qualified = False
      break
  if qualified:
    qualified_roles.append(x['role'])

qualified_roles
```

```python
# performs same function as code above but with less lines
qualified_roles = []

for x in job_roles: # x stores the roles
  if all(c in x['skills'] for c in my_skills): # checks if your skill exists in job posting
    qualified_roles.append(x['role']) # job role to qualified_roles list

qualified_roles
```

## Classes

```python
class AneleList:
    def __init__(self):
        """Initialize an empty list."""
        self._items = []

    def add(self, item):
        """Add an item to the end of the list."""
        self._items.append(item)

    def __getitem__(self, index):
        """Retrieve an item by index."""
        return self._items[index]

    def __setitem__(self, index, value):
        """Set an item at a specific index."""
        self._items[index] = value

    def __repr__(self):
        """Return a string representation of the list."""
        return str(self._items)

    def __len__(self):
        """Return the length of the list."""
        return len(self._items)
```
